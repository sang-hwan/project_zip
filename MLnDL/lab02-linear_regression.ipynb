```python
import numpy as np
import tensorflow as tf
```

```python
x_train = [1, 2, 3, 4]
y_train = [0, -1, -2, -3]

tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=1)) # units = output shape, input_dim = input shape

sgd = tf.keras.optimizers.SGD(lr=0.1) # sgd = standard gradient descendent, lr = learning rate
tf.model.compile(loss='mse', optimizer=sgd) # mse = mean squared error, 1/m*sig(y'-y)^2

tf.model.summary() # print summary of the model to the terminal

tf.model.fit(x_train, y_train, epochs=200) # fit() executes training

y_predict = tf.model.predict(np.array([5, 4]))
print(y_predict)
```

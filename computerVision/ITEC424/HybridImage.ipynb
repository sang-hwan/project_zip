```python
import scipy
from matplotlib import pyplot as plt
from skimage import io
from skimage.transform import resize, downscale_local_mean

image1 = io.imread('/content/Elon_Musk_Royal_Society.jpg')
image2 = io.imread('/content/KakaoTalk_20220321_142546448.jpg')
image1 = resize(image1, (256, 256))
image2 = resize(image2, (256, 256))
low_image = scipy.ndimage.gaussian_filter(image1, 5)
high_image = image2 - scipy.ndimage.gaussian_filter(image2, 5)
hybrid_image = low_image + high_image

plt.imshow(hybrid_image)
plt.rcParams["figure.figsize"] = (1, 1)
plt.show()
```

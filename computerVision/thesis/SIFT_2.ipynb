```python
import cv2
```

```python
img1 = cv2.imread('/content/drive/MyDrive/ComputerVision/Image/logo1.jpg')
img2 = cv2.imread('/content/drive/MyDrive/ComputerVision/Image/logo2.jpg')
```

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
sift = cv2.xfeatures2d.SIFT_create()
```

```python
keypoints_1, descriptors_1 = sift.detectAndCompute(img1, None)
keypoints_2, descriptors_2 = sift.detectAndCompute(img2, None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)
matches = bf.match(descriptors_1, descriptors_2)
```

```python
matches = sorted(matches, key = lambda x:x.distance)
matched_img = cv2.drawMatches(img1, keypoints_1, img2, keypoints_2, matches[:30], img2, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(matched_img)
```
